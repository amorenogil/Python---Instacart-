**Q6. Create a spending flag for each user based on the average price across all their orders using the following criteria:
 If the mean of the prices of products purchased by a user is lower than 10, the flag them as a "Low spender"
 IF the mean of the prices of products purchased by a user is higher or equal to 10, then flag them as "High Spender"**

```python
# Agreggating an "average price" column with transform function
df_def['average_price'] = df_def.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Creating the spending flags
df_def.loc[df_def['average_price'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
df_def.loc[df_def['average_price'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
#chechking the results
df_def['spending_flag'].value_counts(dropna = False)
```

```python
df_def[['user_id', 'spending_flag', 'average_price']].tail(60)
```

**Q7. Create an order frequency flag that marks the regularity of a user's ordering behavior according to the median in the "days_since_prior_order" column. The criteria for the flag should be as follows**: 

1. If the median is higher than 20, then the customer should be labeled a "Non-frequent customer"
2. If the median is higher than 10 and lower or equal to 20, then the customer should be labeled a "Regular customer"
3. If the median is lower than or equal to 10, then the customer should be labeled a "Frequent customer". 

```python
# Agreggating an "order_frequency column" with transform function
df_def['order_frequency'] = df_def.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Creating the frequency flag
df_def.loc[df_def['order_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df_def.loc[(df_def['order_frequency'] <=20 ) & (df_def['order_frequency'] > 10),'frequency_flag'] = 'Regular customer'
```

```python
df_def.loc[df_def['order_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
df_def['frequency_flag'].value_counts(dropna = False)
```

```python
df_def[['user_id', 'order_frequency', 'frequency_flag']].head(40)
```
