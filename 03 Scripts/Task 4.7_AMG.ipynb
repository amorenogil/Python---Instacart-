# 03. Practice of exercise 4.7 deriving New variables

```python
#Defining a new function to classify the products in three price ranges using an If-statement
def price_label(row): 
    if row['prices']<=5:
        return 'Low-range product'
    elif(row['prices']>5) and (row['prices']<=15):
        return 'Mid-range product'
    elif row['prices']>15:
        return 'High Range'
    else: return 'Not enough data'
```

```python
df_test['price_range'] = df_test.apply(price_label, axis=1)
```

```python
# Checking the values of each defined category
df_test['price_range'].value_counts(dropna=False)
```

```python
#checking the most expensive product
df_test['prices'].max()
```

**Q4. Identify the busiest hour of the day. Present the results in periods of time labeled "most orders", "average orders" 
and "fewest orders". Create a new column containing these labels called "busiest_period_day"**

```python
#checking the busiest hours of the day
df_def['time_of_the_order'].value_counts(dropna = False)
```

```python
#creating a new dataframe "sales_count" with the data 

df_sales_counts = df_def['time_of_the_order'].value_counts(dropna = False).rename_axis('hour').reset_index(name = 'sales_count')
```

```python
df_sales_counts
```

```python
#Creating a list with 25th, 50th and 75 quartile values of sales
quartile_orders = list(df_def['time_of_the_order'].value_counts().quantile([0.25, 0.5, 0.75]))
```

```python
quartile_orders
```

**Creating a list for each category: "most orders", "average orders" and "fewest orders"** 

```python
#Most orders list
mst_ords = list(df_sales_counts.loc[df_sales_counts['sales_count'] >= quartile_orders[2]]['hour'])
```

```python
# Average orders list
avg_ords = list(df_sales_counts.loc[(df_sales_counts['sales_count'] > quartile_orders [0]) & (df_sales_counts['sales_count'] < quartile_orders [2])]['hour'])
```

```python
#Fewest order list
fewst_ords = list(df_sales_counts.loc[df_sales_counts['sales_count'] <= quartile_orders[0]]['hour'])
```

**Creating three indexes for the three different labels: Most, Average and Fewst orders**

```python
#index for most orders 
index_mst = df_def['time_of_the_order'].isin(mst_ords)
```

```python
#index for average orders
index_avg = df_def['time_of_the_order'].isin(avg_ords)
```

```python
#index for fewest orders
index_fewst = df_def['time_of_the_order'].isin(fewst_ords)
```

**Creating the new column "busiest_period_of day" and filling it with the three created categories**

```python
df_def.loc[index_mst, 'busiest_period_of_day'] = 'most orders'
```

```python
df_def.loc[index_avg, 'busiest_period_of_day'] = 'average orders'
```

```python
df_def.loc[index_fewst, 'busiest_period_of_day'] = 'fewest orders'
```
