# 03. Practice of Exercise 4.8: Grouping Data & Aggregating Variables

```python
# Creating a subset
df_test = ords_prods_merge [:1000000]
```

```python
df_test.head()
```

```python
# Grouping data by product name
df_test.groupby('product_name')
```

```python
#Grouping data by department id and average number of orders
df_test.groupby('department_id').agg({'order_number': ['mean']})
```

```python
#Alternative method using the .mean function
df_test.groupby('department_id')['order_number'].mean()
```

```python
# Performing multiple agreggations
df_test.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
# Agreggating data with transform to know the number of orders of each user
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# Creating a "loyalty flag" to classify the customers according to the number of products purchased
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10 ),'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order']<=10 , 'loyalty_flag'] = 'New customer'
```

```python
# Checking the results
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```
